## **Контрольні запитання**



1. **У чому відмінність інтерпретатора від компілятора? Python – компілятор чи інтерпретатор?**

    - компілятор перетворює код у мову низького рівня, та після компіляції ми отримаємо вихідний файл який можна буде запустити без аналізу вихідного коду.

    - інтерпретатор виконує перетворення поступово та не створює вихідний файл. Python являється інтерпретатором.

2. **Що таке IDE PyCharm?**
    - середовище розробки для мови Python.    
<br>
    

3. **Що таке система контроля версій (СКВ) і для чого вона призначена? У чому принципова відмінність СКВ SVN та GIT?**

    - СКВ - це програмне забезпечення, що дозволяє відстежувати зміни в файлах.

    - SVN - це централізована система контролю версій. Файли зберігаються на сервері, а користувачі працюють з іх копіями які отримують з сервера.

    - Git - це розподілена система контролю версій. Кожен клон містить повну історію та повний набір файлів.

4. **Що відбувається при виконанні команди git add .?**
    - Оновлює редаговані файли, та додає нові.  
    <br>

5. **Що відбувається при виконанні команди git commit?**
    - Ця команда створює новий комміт.
    <br><br>

6. **Якою Git-командою виконується розміщення локальної копії Git-репозиторію проєкта на віддаленому, що знаходиться на GitHub?**
    - git push
